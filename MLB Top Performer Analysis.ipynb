```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ruth = pd.read_csv('babe-ruth-adv.csv')
ruth=ruth.drop_duplicates(['Year'], keep='first')
ruth=ruth.head(22)
ruth['CWAR']=ruth['WAR'].cumsum()
plt.figure()
ax = plt.subplot(111)
ax.plot(ruth['Age'].astype(int),ruth['CWAR'].astype(float),'-',color='r')

johnson = pd.read_csv('walter-johnson-adv.csv')
johnson=johnson.drop_duplicates(['Year'], keep='first')
johnson = johnson.head(21)
johnson['CWAR']=johnson['WAR'].cumsum()
ax.plot(johnson['Age'].astype(int),johnson['CWAR'].astype(float),'-',color='blue')

bonds = pd.read_csv('barry-bonds-adv.csv')
bonds=bonds.drop_duplicates(['Year'], keep='first')
bonds = bonds.head(22)
bonds['CWAR']=bonds['WAR'].cumsum()
plt.plot(bonds['Age'].astype(int),bonds['CWAR'].astype(float),'--',color='black')

leg = plt.gca().legend(['Babe Ruth','Walter Johnson', 'Barry Bonds', 'Craig Biggio'],frameon=False)
for text in leg.get_texts():
    text.set_color("dimgray")
plt.gca().set_ylabel('Cumulative WAR').set_color("dimgray")
plt.gca().set_xlabel('Age').set_color("dimgray")
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelcolor='dimgray')
plt.tick_params(axis='y', which='both', left=False, right=False, labelcolor='dimgray')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.title('MLB All-Time Wins Above Replacement (WAR) Leaders', color='dimgray')
ax.spines['bottom'].set_color('dimgray')
ax.spines['left'].set_color('dimgray')

plt.axhline(y=63.2, color='gray', linestyle='-.')
plt.text(30, 53, "WAR for Average Hall of Famer", color='gray')
```
